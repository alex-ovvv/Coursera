## Индексация и селекция

```python
frame = pd.read_csv("dataset.tsv", sep = "\t", header=0 )
```

```python
frame
```

```python
#тип объектов каждого столбца
frame.dtypes
```

```python
#это позволяет заменить тип object у столбца Birth на тип datetime - тип время, что позволит использовать специфичные функции
frame.Birth = frame.Birth.apply(pd.to_datetime)
```

```python
frame
```

```python
frame.dtypes
```

```python
#функция позволяет получить мини-инфу по всей таблице
frame.info()
```

```python
#функция позволяет заполнить все пропуски (нулевые значения) на какое-то значение
#inplace=True - говорит, что эти изменения нужно сразу внести в файл, а не создавать его временную копию
frame.fillna("разнорабочий", inplace = True)
```

```python
frame
```

```python
#функция выводит конкретные столбец, который мы запросили
frame.Position
```

```python
#если же мы хотим вывести его в виде таблицы, то функция будет другой
frame[["Position"]]
```

```python
frame[["Position", "Name"]]
```

```python
#функция выводит первые n-строк таблицы
frame.head(3)
```

```python
#другой способ вывода строк
frame[1:3] #первые три строки
frame[-3:] #последние три строки
```

```python
#функция вывода определенных строк и столбцов таблицы
frame.loc[[1,3,5], ["Name", "City"]]
```

```python
#функция вывода определенных строк и столбцов таблицы, только столбцы по номерам, а не названиями
frame.iloc[[1,3,5], [1,2]]
```

```python
#можно делать и разнообразные логические операции и фильтрации
frame[frame.Birth >= pd.datetime(1985,1,1)] #например, здесь мы оставили только тех, кто старше 1985 года
```

### если логических условий два, например, город и возраст, то синтаксис у такой записи будет следующим
frame[ () & () ] - знак & говорит о том, что мы будем использовать оба условия (пересечение условий)

frame[ () | () ] - знак | говорит о том, что мы хотим использовать или или (хотя б одно условие)

```python
frame[(frame.Birth >= pd.datetime(1985,1,1))&(frame.City != "Москва")]
```

```python
frame[(frame.Birth >= pd.datetime(1985,1,1))|(frame.City == "Волгоград")]
```
